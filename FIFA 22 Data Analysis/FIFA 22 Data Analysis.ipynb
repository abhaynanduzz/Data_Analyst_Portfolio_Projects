```python
#plot the relation between playerpotential and wages


plt.figure(figsize=(7,5))
sns.scatterplot(x=df.potential,y=df.wage_eur)
plt.xlabel('Potential')
plt.ylabel('Wage Eur')
plt.title('Potential and Wage',fontsize=18)
plt.show()
```

```python
#plot the relation between playerpotential and wages, filtered by International reputation


plt.figure(figsize=(7,5))
sns.scatterplot(x=df.potential,y=df.wage_eur,hue=df.international_reputation)
plt.xlabel('Potential')
plt.ylabel('Wage Eur')
plt.title('Potential and Wage',fontsize=18)
plt.show()
```

```python
#Relation between player potential and values
plt.figure(figsize=(7,5))
sns.scatterplot(x=df.potential,y=df.value_eur)
plt.xlabel('Potential')
plt.ylabel('Value in EUR')
plt.title('Potential and Value in EUR',fontsize=18)
plt.show()
```

```python
 #the relationship between player reputation and value
    
    
sns.scatterplot(x=df.international_reputation,y=df.value_eur)
plt.xlabel('International Reputation')
plt.ylabel('Value in EUR')
plt.title('Reputation and Value in EUR',fontsize=18)
plt.show()
    
```

```python
#plot the relationship between player reputation and wages

sns.scatterplot(x=df.height_cm,y=df.potential)
plt.xlabel(' Height')
plt.ylabel('Potential')
plt.title('Relationship between Height and Potential',fontsize=18)
plt.show()
    


```

```python
nation_name=' '.join(n for n in df.nationality_name)
plt.figure(figsize=(10,10))
wc = WordCloud().generate(nation_name) 
plt.imshow(wc, interpolation='bilinear') 
plt.axis('off') 
plt.show()
```

```python
def plot_most_common(category):
    bar_plot = dict(Counter(df[category].values).most_common(5))
    plt.bar(*zip(*bar_plot.items()))
    plt.show()
```

```python
# What are the most common player positions?

fig, ax = plt.subplots(figsize = (8,5))
plot_most_common('player_positions')
plt.show()
```
